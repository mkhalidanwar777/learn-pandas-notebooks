# merge function

```python
import numpy as np
import pandas as pd

delivery=pd.read_csv('deliveries.csv')
delivery.head(2)
```

```python
import numpy as np
import pandas as pd

match=pd.read_csv('matches - matches.csv')
match.head(2)
```

# now merge both file 
## with the help of match_id and id column 

```python
delivery.merge(match,left_on='match_id',right_on='id').head(2)
```

# now we find in every season hieghest runs player
### step by step
#
# step 1

```python
new=delivery.merge(match,left_on='match_id',right_on='id')
new.groupby(['season','batsman'])['batsman_runs'].sum()
```

# step 2

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False)
```

# step 3
### this series data convert datafrema ( .reset_index() )

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```

# step 4
### each season i want only one player ( drop_duplicates(subset='season' , keep='first')

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first')
```

# step 5 
### again sorting 

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
```

# step 6
### i want only season and batsman name column 

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batsman']]
```
