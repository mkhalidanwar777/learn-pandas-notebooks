# Sorting in Series Data

```python
import numpy as np
import pandas as pd

data=pd.read_csv('battingAvg.csv')
print(data['Team Name'].value_counts())
print()
print('Now sorting')

data['Team Name'].value_counts().sort_values()
```

# Descending Order

```python
import numpy as np
import pandas as pd

data=pd.read_csv('battingAvg.csv')
print(data['Team Name'].value_counts().sort_values())
print()
print('descending Order')
print()
data['Team Name'].value_counts().sort_values(ascending=False)
```

# Sorting in Datafrema Data
