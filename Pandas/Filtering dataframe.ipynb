# Filtering dataframe

# Single condition
## 1. i want those matches which play  Islamabad United

```python
import numpy as np
import pandas as pd

data=pd.read_csv('battingavg.csv')
mask=data['Team Name']=='Islamabad United'
data[mask]
```

# More then one condition 
#
## play in Islamabad United and year is 2023

```python
import numpy as np
import pandas as pd

data=pd.read_csv('battingavg.csv')
mask1=data['Team Name']=='Islamabad United'
mask2=data['Span']>'2023'
data[mask1 & mask2]
```
