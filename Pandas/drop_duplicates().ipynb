## drop_duplicates('Team Name')

```python
import numpy as np
import pandas as pd
data=pd.read_csv('battingAvg.csv')

data.drop_duplicates('Team Name')
```

## Multiply Column drop_duplicates(subset=['Span','Team Name'])

```python
import numpy as np
import pandas as pd
data=pd.read_csv('battingAvg.csv')

data.drop_duplicates(subset=['Span','Team Name'])
```

## keep='last' 

```python
import numpy as np
import pandas as pd
data=pd.read_csv('battingAvg.csv')

data.drop_duplicates('Team Name',keep='last')
```

## Only Two colums

```python
import numpy as np
import pandas as pd
data=pd.read_csv('battingAvg.csv')

data.drop_duplicates('Team Name',keep='last')[['Team Name','Span']]
```

## sort with dates

```python
import numpy as np
import pandas as pd
data=pd.read_csv('battingAvg.csv')

data.drop_duplicates('Team Name',keep='last')[['Team Name','Span']].sort_values('Span')
```
