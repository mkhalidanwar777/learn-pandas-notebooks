# Missing value handle
## 1.dropna()
## 2.fillna()

```python
import pandas as pd

data=pd.read_csv('battingAvg.csv')
data.head()
```

# if any row have one value is missing the dropna() function remove this row

```python
data.dropna().head()
```

# now apply condition on column if this column have missing then remove

```python
data.dropna(subset=['NO','Runs']).head()
```

## in above condition is OR

# 2.fillna()

```python
data.head()
```

```python
data['NO'].fillna( 0 , inplace=True)
data.head()
```

```python
data['Runs'].fillna( "NO" , inplace=True)
data.head()
```

```python
data['Team Name'].fillna( " No value " , inplace=True)
data.head()
```

# using method to set value 

```python
data.head()
```

```python
data['SR'].fillna(method='ffill')
```
